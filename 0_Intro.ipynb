# Introduction - Ray - Anyscale

## What is Ray?

* Ray is a framework for high-performance, resilient, scale-out computation on heterogeneous hardware
* Ray AI Libraries are a set of high-level APIs for accomplishing common large-scale data + compute use cases (e.g., data transformation, model training)

<img src='https://docs.ray.io/en/releases-2.38.0/_images/map-of-ray.svg' width=700 />

## Why Ray?

* Ray's strengths align perfectly with the requirements for high-performance training, inference, and data processing with large AI models
* The Ray AI libraries make those tasks straightforward while integrating closely with the industry standard tools like PyTorch and Lightning.

## Why Anyscale?

Ray is great open-source software. But Ray is focused on core capabilities for AI computation. 

Anyscale adds infrastructure management, enterprise-friendly features, and technical extensions that make Ray easy to deploy and operate.

## Anyscale model

__Infrastructure layer__
* Cloud
* Compute config ("hardware env")
* Cluster environment ("software env") based on Ray + Anyscale
    * Container images
    * Versioning
* Cluster
* Storage
    * Persistence scopes: instance, cluster, user, organization

__Service layer__
* Workspace
* Jobs
* Anyscale Services (based on Ray Serve)
* Schedules

__Admin/operational/security features__
* Automation-friendly user and resource management
* SSO support
* Granular access controls
* Audit logs

__Additional technical features (partial list)__
- Specialized Ray Data ingest support for video, audio, Databricks, and Snowflake sources
- Out of the box integration with NFS cloud storage
- Support for Non-Volatile Memory Express (NVMe) as a cutting-edge interface to access SSD storage volumes
    -   Offload data from costly RAM storage to SSDs, improve perf compared to other local storage
